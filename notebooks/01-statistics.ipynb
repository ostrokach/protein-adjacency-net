# Create summary Parquet files

## `pc_identity`

```python
ds = spark.sql(f"""\
SELECT __index_level_0__, database_id, pc_identity
FROM parquet.`{ADJACENCY_MATRIX_PATH}`
""")
```

```python
pc_identify_df = pq.read_table(f'{NOTEBOOK_NAME}/pc_identity.parquet').to_pandas()
```

```python
pc_identify_df.head()
```

```python
hist, bin_edges = np.histogram(pc_identify_df['pc_identity'].values, bins=10, range=(0, 100))
```

```python
fg, axs = plt.subplots(figsize=(10, 6))

plt.bar([(bin_edges[i] + bin_edges[i + 1]) / 2 for i in range(len(hist))], hist, 8)

plt.xlabel("Sequence identity")
plt.ylabel("Number of domains")
plt.xlim(0, 100)
```
