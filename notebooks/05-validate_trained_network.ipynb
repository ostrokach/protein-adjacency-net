```python
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio import SeqIO, AlignIO
```

```python
seqs = list(AlignIO.read(NOTEBOOK_PATH.joinpath('hhblits_reducedQT_1456673.fasta').as_posix(), 'fasta'))
```

```python
aa_idxs = [i for i, aa in enumerate(seqs[0].seq) if aa != '-']

new_seqs = []
for seq in seqs[1:]:
    new_seq = SeqRecord(
        seq=Seq(''.join(seq.seq[i] for i in aa_idxs)),
        id=seq.id, name=seq.name, description=seq.description,
    )
    new_seqs.append(new_seq)
```

```python
SeqIO.write(
    new_seqs,
    NOTEBOOK_PATH.joinpath('hhblits_reducedQT_1456673_nogaps.fasta').as_posix(),
    'fasta'
)
```
